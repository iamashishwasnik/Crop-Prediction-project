## Treating skewness

```python
from sklearn.preprocessing import PowerTransformer
```

```python
pt = PowerTransformer()
```

```python
columns_to_scale = ['rainfall', 'temperature', 'Area_in_hectares', 'Production_in_tons', 'Yield_ton_per_hec']

df[columns_to_scale] = pt.fit_transform(df[columns_to_scale])
```

```python
for i in columns_to_scale:
    sns.kdeplot(data=df,x=i)
    plt.show()
```

## Label encoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
LE=LabelEncoder()
```

```python
categorical_columns=['State_Name','Crop_Type','Crop']
```

```python
for column in categorical_columns:
    df[column] = LE.fit_transform(df[column])
```

## Split the data for training & testing

```python
from sklearn.model_selection import train_test_split
```

```python
X = df[['State_Name', 'Crop_Type', 'Crop', 'N', 'P', 'K', 'pH', 'rainfall', 'temperature', 'Area_in_hectares']]
y = df['Yield_ton_per_hec']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

## LinearRegression

```python
from sklearn.linear_model import LinearRegression,Lasso
```

```python
LR=LinearRegression().fit(X_train,y_train)
```

```python
LR.score(X_train,y_train)
```

```python
y_pred=LR.predict(X_test)
```

```python
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
```

```python
mse=mean_squared_error(y_test,y_pred)
mae=mean_absolute_error(y_test,y_pred)
r2_score(y_test,y_pred)

print("r2_score:            ",(r2_score(y_test,y_pred)))
print("mean_squared_error:  ",(mean_squared_error(y_test,y_pred)))
print("mean_absolute_error: ",(mean_absolute_error(y_test,y_pred)))
```

```python
residual=y_test-y_pred #Residuals are the differences between the observed values and the values predicted by model
```

## DecisionTreeRegressor

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
DTR=DecisionTreeRegressor(random_state=0).fit(X_train,y_train)
```

```python
DTR.score(X_train,y_train)
```

```python
y_pred=DTR.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```

```python
print("r2_score:            ",(r2_score(y_test,y_pred)))
print("mean_squared_error:  ",(mean_squared_error(y_test,y_pred)))
print("mean_absolute_error: ",(mean_absolute_error(y_test,y_pred)))
```
